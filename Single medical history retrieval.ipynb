```python
import json
import pandas as pd

# Sample FHIR message (as a dictionary)
fhir_message = {
    "resourceType": "Patient",
    "extension": [
        {
            "url": "http://hl7.org/fhir/StructureDefinition/patient-birthPlace",
            "valueAddress": {
                "city": "Barnstable",
                "state": "Massachusetts",
                "country": "US"
            }
        }
    ],
    "identifier": [
        {
            "system": "https://github.com/synthetichealth/synthea",
            "value": "8c95253e-8ee8-9ae8-6d40-021d702dc78e"
        },
        {
            "type": {
                "coding": [
                    {
                        "system": "http://terminology.hl7.org/CodeSystem/v2-0203",
                        "code": "MR",
                        "display": "Medical Record Number"
                    }
                ],
                "text": "Medical Record Number"
            },
            "system": "http://hospital.smarthealthit.org",
            "value": "8c95253e-8ee8-9ae8-6d40-021d702dc78e"
        },
        {
            "type": {
                "coding": [
                    {
                        "system": "http://terminology.hl7.org/CodeSystem/v2-0203",
                        "code": "SS",
                        "display": "Social Security Number"
                    }
                ],
                "text": "Social Security Number"
            },
            "system": "http://hl7.org/fhir/sid/us-ssn",
            "value": "999-86-2571"
        }
    ],
    "name": [
        {
            "use": "official",
            "family": "Dickens475",
            "given": ["Aaron697"],
            "prefix": ["Mr."]
        }
    ],
    "telecom": [
        {
            "system": "phone",
            "value": "555-152-6034",
            "use": "home"
        }
    ],
    "gender": "male",
    "birthDate": "1944-08-28",
    "deceasedDateTime": "1998-08-15T13:05:53+01:00",
    "address": [
        {
            "line": ["859 Altenwerth Run Unit 88"],
            "city": "Charlton",
            "state": "MA",
            "country": "US"
        }
    ],
    "maritalStatus": {
        "coding": [
            {
                "system": "http://terminology.hl7.org/CodeSystem/v3-MaritalStatus",
                "code": "M",
                "display": "Married"
            }
        ],
        "text": "Married"
    },
    "communication": [
        {
            "language": {
                "coding": [
                    {
                        "system": "urn:ietf:bcp:47",
                        "code": "en-US",
                        "display": "English"
                    }
                ],
                "text": "English"
            }
        }
    ]
}


def extract_patient_data(fhir_message):
    patient_data = {
        "ID": fhir_message.get("identifier", [{}])[0].get("value"),
        "Name": " ".join(fhir_message.get("name", [{}])[0].get("given", [])) + " " + fhir_message.get("name", [{}])[0].get("family", ""),
        "Gender": fhir_message.get("gender"),
        "Birth Date": fhir_message.get("birthDate"),
        "Deceased Date": fhir_message.get("deceasedDateTime"),
        "Phone": fhir_message.get("telecom", [{}])[0].get("value"),
        "Address": ", ".join(fhir_message.get("address", [{}])[0].get("line", [])) + ", " + fhir_message.get("address", [{}])[0].get("city") + ", " + fhir_message.get("address", [{}])[0].get("state") + ", " + fhir_message.get("address", [{}])[0].get("country"),
        "Marital Status": fhir_message.get("maritalStatus", {}).get("text"),
        "Language": fhir_message.get("communication", [{}])[0].get("language", {}).get("text"),
        "Birth Place": fhir_message.get("extension", [{}])[0].get("valueAddress", {}).get("city") + ", " + fhir_message.get("extension", [{}])[0].get("valueAddress", {}).get("state") + ", " + fhir_message.get("extension", [{}])[0].get("valueAddress", {}).get("country")
    }
    return patient_data


patient_data = extract_patient_data(fhir_message)
df = pd.DataFrame([patient_data])
df.to_csv('patient_data.csv', index=False)
print("Patient data has been transformed and saved to patient_data.csv")
```
